```python
import pandas as pd
```

```python
crashes = pd.read_csv('./crashes_cleaned.csv')
people = pd.read_csv('./people_cleaned.csv')
vehicles = pd.read_csv('./vehicles_cleaned.csv')
```

```python
data = crashes.merge(people, on='CRASH_RECORD_ID', how='inner').merge(vehicles, on='CRASH_RECORD_ID', how='inner')
```
